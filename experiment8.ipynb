```python
"""1.	Add few names, one name in each row, in “name.txt file”. 
a.	Count no of names 
b.	Count all names starting with vowel 
c.	Find longest name 
"""

def count_names(file_path):
    with open(file_path, 'r') as file:
        names = file.read().splitlines()
        return len(names)
def count_names_starting_with_vowel(file_path):
    vowels = 'AEIOUaeiou'
    count = 0
    with open(file_path, 'r') as file:
        names = file.read().splitlines()
        for name in names:
            if name and name[0] in vowels:
                count += 1
    return count
def find_longest_name(file_path):
    longest_name = ""
    with open(file_path, 'r') as file:
        names = file.read().splitlines()
        for name in names:
            if len(name) > len(longest_name):
                longest_name = name
    return longest_name
if __name__ == "__main__":
    file_path = 'name.txt'
    total_names = count_names(file_path)
    names_starting_with_vowel = count_names_starting_with_vowel(file_path)
    longest_name = find_longest_name(file_path)
    
    print(f"Total number of names: {total_names}")
    print(f"Number of names starting with a vowel: {names_starting_with_vowel}")
    print(f"Longest name: {longest_name}")
    

```
