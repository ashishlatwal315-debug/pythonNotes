```python
class MatrixArray:

    def __init__(self, matrix1, matrix2):
        self.matrix1 = matrix1
        self.matrix2 = matrix2

    def display(self, matrix):
        for row in matrix:
            print(row)

    def transpose(self, matrix):
        trans = []
        for j in range(len(matrix[0])):
            row = []
            for i in range(len(matrix)):
                row.append(matrix[i][j])
            trans.append(row)
        return trans

    def cross_multiply(self):
        result = []
        for i in range(len(self.matrix1)):
            row = []
            for j in range(len(self.matrix2[0])):
                total = 0
                for k in range(len(self.matrix2)):
                    total += self.matrix1[i][k] * self.matrix2[k][j]
                row.append(total)
            result.append(row)
        return result


# -------- USER INPUT --------

# Matrix 1
r1 = int(input("Enter rows of matrix 1: "))
c1 = int(input("Enter columns of matrix 1: "))
matrix1 = []

print("Enter matrix 1:")
for i in range(r1):
    matrix1.append(list(map(int, input().split())))

# Matrix 2
r2 = int(input("Enter rows of matrix 2: "))
c2 = int(input("Enter columns of matrix 2: "))
matrix2 = []

print("Enter matrix 2:")
for i in range(r2):
    matrix2.append(list(map(int, input().split())))

obj = MatrixArray(matrix1, matrix2)

print("\nTranspose of Matrix 1:")
obj.display(obj.transpose(matrix1))

print("\nTranspose of Matrix 2:")
obj.display(obj.transpose(matrix2))

print("\nCross Multiplication:")
obj.display(obj.cross_multiply())
```
