```python
# if we make any changes using instance it doesnot cahange value inside the class variable busing class method we change vthw value 
# self work on a object or a instace but class method work in class
class hello():
    raise_ = 2
    def __init__(self , fname,lname,pay):
        self.fname = fname 
        self.lname= lname 
        self.pay = pay
    def new(self):
        print(f"{self.fname} {self.lname}")
    def raise_pay(self):
        self.pay = self.pay*2

    @classmethod
    def new_raise(cls,amount,lname):
        cls.raise_ = amount
        cls.fname = lname

s1 = hello("ashish","latwal",20000)# instance
hello.new_raise(5,) # or hello.raie_ = 5 or s1.new_raise(5)


print(s1.raise_)           
print(hello.raise_)
```
