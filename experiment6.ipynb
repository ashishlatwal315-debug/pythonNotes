```python
#1.	Scan n values in range 0-3 and print the number of times each value has occurred

n = int(input("Enter the number of values: "))
values = []
for i in range(n):
    value = int(input("Enter a value (0-3): "))
    if 0 <= value <= 3:
        values.append(value)
    else:
        print("Invalid input. Please enter a value between 0 and 3.")


occurrences = [0, 0, 0, 0]  # To count occurrences of 0, 1, 2, and 3
for value in values:
    occurrences[value] += 1
print("Value occurrences:")
for i in range(4):
    print(f"{i}: {occurrences[i]} times")
        
```

```python
#2.	Create a tuple to store n numeric values and find the average of all values

n = int(input("Enter the number of values: "))
values = []
for i in range(n):
    value = float(input("Enter a numeric value: "))
    values.append(value)
values_tuple = tuple(values)
average = sum(values_tuple) / len(values_tuple)
print(f"The average of the values is: {average}")
```

```python
"""3.	WAP to input a list of scores for N students in a list data type. Find the score of the runner-up and print the output. 
Sample Input 
N = 5 
Scores= 2 3 6 6 5 
Sample output 
5 
"""

n = int(input("Enter the number of students: "))
scores = []
for i in range(n):
    score = int(input("Enter the score for student {}: ".format(i + 1)))
    scores.append(score)
unique_scores = list(set(scores))  # Remove duplicates
unique_scores.sort()  # Sort the scores
if len(unique_scores) < 2:
    print("Not enough unique scores to determine a runner-up.")
else:
    runner_up_score = unique_scores[-2]  # The second highest score
    print(f"The runner-up score is: {runner_up_score}")

```

```python
"""4.	Create a dictionary of n persons where key is name and value is city.  
a.	Display all names 
b.	Display all city names 
c.	Display student names and cities of all students. 
d.	Count the number of students in each city
"""

n = int(input("Enter the number of persons: "))
persons = {}
for i in range(n):
    name = input("Enter the name of person {}: ".format(i + 1))
    city = input("Enter the city of person {}: ".format(i + 1))
    persons[name] = city
# a. Display all names
print("Names of all persons:")
for name in persons.keys():
    print(name)
# b. Display all city names
print("City names of all persons:")
for city in set(persons.values()):
    print(city)

# c. Display student names and cities of all students
print("Names and cities of all persons:")
for name, city in persons.items():
    print(f"{name} lives in {city}")
# d. Count the number of students in each city
city_count = {}
for city in persons.values():
    if city in city_count:
        city_count[city] += 1
    else:
        city_count[city] = 1
print("Number of persons in each city:")
for city, count in city_count.items():
    print(f"{city}: {count} persons")

    
```

```python
#5.	Store details of n movies in a dictionary by taking input from the user. Each movie must store details like name,  year, director name, production cost, collection made (earning) & perform the following :- 
"""a.	print all movie details 
b.	display name of movies released before 2015 
c.	print movies that made a profit
d.	print movies directed by a particular director
"""

n = int(input("Enter the number of movies: "))
movies = {}
for i in range(n):
    name = input("Enter the name of movie {}: ".format(i + 1))
    year = int(input("Enter the release year of movie {}: ".format(i + 1)))
    director = input("Enter the director name of movie {}: ".format(i + 1))
    production_cost = float(input("Enter the production cost of movie {}: ".format(i + 1)))
    collection_made = float(input("Enter the collection made (earning) of movie {}: ".format(i + 1)))
    movies[name] = {
        "year": year,
        "director": director,
        "production_cost": production_cost,
        "collection_made": collection_made
    }
print("Movie details:")
for name, details in movies.items():
    print(f"Name: {name}, Year: {details['year']}, Director: {details['director']}, Production Cost: {details['production_cost']}, Collection Made: {details['collection_made']}")
print("Movies released before 2015:")
for name, details in movies.items():
    if details["year"] < 2015:
        print(name)
print("Movies that made a profit:")
for name, details in movies.items():
    if details["collection_made"] > details["production_cost"]:
        print(name)
director_name = input("Enter the director name to search for movies: ")
print(f"Movies directed by {director_name}:")
for name, details in movies.items():
    if details["director"] == director_name:
        print(name)

        

```
