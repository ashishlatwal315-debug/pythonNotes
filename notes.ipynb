# Strings

- String is a sequence of characters.
- Strings are IMMUTABLE (cannot be changed after creation).
- Created using single, double or triple quotes.

Example:
'hello'
"hello"
"""hello"""

```python
# String creation
s = "Hello World"

# Indexing (0-based)
print(s[0])      # H
print(s[-1])     # d

# Slicing
print(s[0:5])    # Hello
print(s[6:])     # World

# Operators
print(s + " !!!")   # Concatenation
print(s * 2)        # Repetition

# Methods
print(s.lower())
print(s.upper())
print(s.split())
```

# Lists

- List is a collection of items.
- Lists are MUTABLE (can be changed).
- Created using square brackets [].
- Can store different data types.

```python
# List creation
lst = [10, 20, 30, 40]

# Access elements
print(lst[1])

# Modify element
lst[1] = 25
print(lst)

# Add elements
lst.append(50)
lst.insert(1, 15)

# Remove elements
lst.remove(30)
lst.pop()

print(lst)

# List comprehension
squares = [x*x for x in range(5)]
print(squares)
```

# Tuples

- Tuple is similar to list.
- Tuples are IMMUTABLE.
- Created using parentheses ().
- Faster than lists.

```python
# Tuple creation
t = (1, 2, 3, 4)

print(t[0])

# Tuple unpacking
a, b, c, d = t
print(a, b)

# Count and index methods
print(t.count(2))
print(t.index(3))
```

# Sets

- Set is an unordered collection.
- No duplicate values allowed.
- Created using {}.
- Supports mathematical operations.

```python
# Set creation
s = {1, 2, 3, 3, 4}
print(s)   # duplicates removed

a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)   # Union
print(a & b)   # Intersection
print(a - b)   # Difference
```

# Dictionaries

- Stores data in key-value pairs.
- Created using {}.
- Keys must be unique.
- Mutable data type.

```python
# Dictionary creation
student = {
    "name": "Ashish",
    "age": 21,
    "course": "BCA"
}

# Access value
print(student["name"])

# Add / Update
student["age"] = 22
student["city"] = "Delhi"

# Remove
student.pop("city")

print(student)

# Loop
for key, value in student.items():
    print(key, value)
```

# Functions

- Function is a block of reusable code.
- Defined using def keyword.
- Can take parameters.
- Can return value.

```python
# Simple function
def greet(name):
    return "Hello " + name

print(greet("Ashish"))

# Default parameter
def add(a, b=10):
    return a + b

print(add(5))
print(add(5, 20))
```

# Lambda Function

- Anonymous function.
- Written in one line.
- Syntax: lambda arguments: expression

```python
# Lambda
square = lambda x: x * x
print(square(5))

# Map
numbers = [1, 2, 3, 4]
result = list(map(lambda x: x*2, numbers))
print(result)

# Filter
even = list(filter(lambda x: x % 2 == 0, numbers))
print(even)
```
