 # *Que- What is interactive programming in python?

 # *Ans- Interactive programming is a method where you engage with the Python interpreter in real-time.

   # *Execution*: You can run programming code directly from the terminal or command prompt.

   # *Feedback*: The output of the code is instant upon pressing enter.

   # *Efficiency*: You do not need a helper command to run the script; you simply type the code and see the result immediately.

 # *Que- what is script programming in python?

 # *Ans- Script programming involves writing code in a separate file and executing it all at once.

   # *Process*: You write your code in a file and then submit the script to the interpreter/compiler.

   # *Execution*: Unlike interactive mode, you use a helper command to run the code.

   # *Command*: In the case of Python, the specific terminal command used is:

   # python filename.py

# ==========================================
# DYNAMIC TYPING IN PYTHON
# ==========================================

# Python is a "Dynamically Typed" language.
# 1. Variables are declared during runtime.
# 2. Data types are NOT explicitly specified (no need to say int or str).
# 3. The same variable can change types based on the value assigned.

```python
a = 20          
print(type(a))   

a = 20 / 100 
print(type(a))
```

# ==========================================
# IMMUTABLE DATA TYPES
# ==========================================

# Definition: Data types where values cannot be changed once assigned.
# If you try to change a specific part (like a character in a string), 
# Python will throw a TypeError.

# List of Immutable Types:
# 1. Integer (int)
# 2. Floating Point (float)
# 3. Boolean (bool)
# 4. String (str)
# 5. Frozenset
# 6. Tuples

# String Immutability Example:

```python
a = "abcd"
a[1] = "v"
```

# This would cause: TypeError: 'str' object does not support item assignment.

# ==========================================
# THE TUPLE DATA TYPE
# ==========================================

# Definition: A collection of items that is ordered and IMMUTABLE.
# Syntax: Uses parentheses ()

```python
a = (2, 3)
print(a[0])      # Output will be is 2
print(type(a))   # Output: <class 'tuple'>
```

# a[1] = 4       # will give error because You cannot change a value inside a tuple once set.

# =========================================================
# *PYTHON MUTABLE TYPES: "CAN BE CHANGED IN PLACE*"
# =========================================================

# 1. *LISTS* []
# - Ordered, indexed, and flexible.
# - Change an item: list[index] = new_value
# - Methods: .append() (add to end), .pop() (remove), .sort()

```python
my_list = [10, 20, 30]
my_list[0] = 99
print(my_list[0])
```

# 2. *DICTIONARIES* {key: value}
# - Store data in pairs; super fast lookups.
# - Add/Update: dict['key'] = value
# - Delete: del dict['key'] or .pop('key')

```python
my_dict = {"name": "Alex", "score": 85}
my_dict["score"] = 90 
my_dict["pass"] = True
```

# 3. *SETS* {}
# - Unordered collection of UNIQUE items.
# - No indexes! (Can't do set[0])
# - Methods: .add(), .remove(), .discard()

```python
my_set = {1, 2, 3}
my_set.add(4) 
my_set.add(2)
print(my_set)
```
