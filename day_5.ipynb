```python
                                                          #   encapsulation
class hello:
    def __init__(self,name,roll,age):
        self.name=name
        self.roll = roll
        self.__age= age# MAKING IT PRIVATE AND CONTROLLING ACCESS

    def record(self):
        print(self.age)


s1 = hello("ashish",18,18)#object #An instance is just another name for an object created from a class

# Call method
s1.record()
                                            # OUTPUT
                             # 'hello' object has no attribute 'age'
```

```python
                        # HOW TO GET VALUE IN ENCAPSULATION
# Encapsulation is the process of making attributes (data) private OR restrict access to certain atributes  
# and providing access to them through public methods
class hello:
    def __init__(self,name,roll,age):
        self.name=name
        self.roll = roll
        self.__age= age

    def get_age(self):
        print(self.__age)


s1 = hello("ashish",18,18)

s1.get_age()
```

```python
#                                   # abstarction
# Abstraction is the process of hiding implementation details and showing only the essential features to the user.
from abc import ABC, abstractmethod
class byy(ABC):    ##### byy is abstract base class we made it by using ABC
    
    @abstractmethod
    def security(self):
        print("this is type of security")
    
class record(byy):
    def security(self):
        print(" my name is ashish")
s1 = record()      #### IT WILL INHERIT THE ABSTRACT CLASS AND DO CERTAIN TASK
s1.security()
# s1.byy() WE CAN NOT USE BYY ABSTRACT CLASS BECAUSE IT IS HIDDDEN FROM THE USER
```

```python
                             # INHERITENCE

class byy():    
    def __init__(self,NAME,CLASS,ROLL_NO):
        self.NAME = NAME
        self.CLASS = CLASS
        self.ROLL_NO = ROLL_NO
    def first(self):
        print(f"my nmae is {self.NAME} from class {self.CLASS} rollno {self.ROLL_NO}")
s1 = byy("ashish",12,18)   
print(s1.NAME,s1.CLASS)
s1.first()

print("                                   after in INHERITENCE")

class new(byy):
    def __init__(self, NAME, CLASS, ROLL_NO,age):
        super().__init__(NAME, CLASS, ROLL_NO)
        self.age = age
    def first(self):
        super().first()
        print(f"my name is{self.NAME} AND MY age is {self.age}")
s1 = byy("ashish",12,18) 
print(s1.NAME,s1.CLASS)
s2 = new("dhruv",19,11,18)
print(s2.NAME,s2.age)

print("                it is inherting method from byy class and we can print it accousoing to object" )

s2.first()
```
