```python
#1 Find a factorial of given number. ()
num = int(input("Enter a number: "))
factorial = 1
if num < 0:
    print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, num + 1):
        factorial *= i
    print("The factorial of", num, "is", factorial)
    
```

```python
#2 Find whether the given number is Armstrong number

num = int(input("Enter a number: "))
order = len(str(num))
sum = 0
temp = num
while temp > 0:
    digit = temp % 10
    sum += digit ** order
    temp //= 10
if num == sum:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
    
```

```python
#3 Print Fibonacci series up to given term

n = int(input("Enter the number of terms: "))
a, b = 0, 1
count = 0
if n <= 0:
    print("Please enter a positive integer")
elif n == 1:
    print("Fibonacci sequence up to", n, ":")
    print(a)
else:
    print("Fibonacci sequence up to", n, ":")
    while count < n:
        print(a)
        a, b = b, a + b
        count += 1
        
```

```python
#4 Write a program to find if given number is prime number or not. 
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, int(num/2) + 1):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
```

```python
#5 Write a program to print sum of digits

num = int(input("Enter a number: "))
sum = 0
while num > 0:
    digit = num % 10
    sum += digit
    num //= 10
print("The sum of the digits is:", sum)
```

```python
#6 Count and print all numbers divisible by 5 or 7 between 1 to 100. 

count = 0
for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i)
        count += 1
print("Total numbers divisible by 5 or 7 between 1 and 100:", count)
```

```python
#7 Convert all lower cases to upper case in a string

string = input("Enter a string: ")
upper_string = ""
for char in string:
    if char.islower():
        upper_string += char.upper()
    else:
        upper_string += char
print("String with lowercase converted to uppercase:", upper_string)
```
