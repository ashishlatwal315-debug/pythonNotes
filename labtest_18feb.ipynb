```python
# code 3 

def missing_number(nums):
    n = len(nums)
    # Use // for integer division to maintain precision and return an int
    total = n * (n + 1) // 2
    return total - sum(nums)
```

```python
# code 4
def add_unique(x, seen=None):
    # Check if a set was provided; if not, create a new one
    if seen is None:
        seen = set()
    
    if x not in seen:
        seen.add(x)
        
    return list(seen)
```
