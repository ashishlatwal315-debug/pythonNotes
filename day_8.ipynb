```python
class hello :
    school_name = "dmga"
    def __init__ (self,name,age):
        self.name = name 
        self.age = age 
s1 = hello("indar",18)
s2 = hello("rohit",19)
print(s1.school_name) #Works because Python looks into class if not found in object
print(s2.school_name)
print(s2.name)



```

```python
class hello :
    school_name = "dmga"
    def __init__ (self,name,age):
        self.name = name 
        self.age = age 
s1 = hello("indar",18)
s2 = hello("rohit",19)
s1.school_name = "ccs"
print(s1.school_name) # it only create new instance variable  for s1 it does namt change in class varible
print(s2.school_name)
print(hello.school_name)




```

```python
class hello :
    school_name = "dmga"
    subject = "it"
    def __init__ (self,name,age):
        self.name = name 
        self.age = age 
    @classmethod                     #Used to modify class variables
    def byy(cls,subject):            #Refers to class (not object)
        cls.subject = subject

s1 = hello("indar",18)
hello.byy("python")
print(hello.subject)                 # change the value of subject it to python





```

```python
class hello :
    school_name = "dmga"
    subject = "it"
    def __init__ (self,name,age):
        self.name = name 
        self.age = age 
    @staticmethod                     
    def byy(name):            
        print(name)
s1 = hello("indar",18)
#print(hello.subject)                 # change the value of subject it to python
hello.byy(s1.name)                    # passing name manually





```
