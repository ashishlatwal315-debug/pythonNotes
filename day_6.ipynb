```python
class byy:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def __add__(self, other):
        return byy(self.a + other.a ,self.b+ other.b, self.c+ other.c)
a = byy(2,3,4)
b = byy(5,6,7)
c = a+b    # # This will call the __add__ method

print(c.a, c.b, c.c)
```

```python
class byy:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def __add__(self, other):
        print("Inside __add__ method")
        print(f"Adding: (a: {self.a}, b: {self.b}, c: {self.c}) + (a: {other.a}, b: {other.b}, c: {other.c})")
        # Create and return a new byy object with summed attributes
        return byy(self.a + other.a, self.b + other.b, self.c + other.c)# now we are crearing a new instance name c
    

# Create two objects
a = byy(2, 3, 4)
b = byy(5, 6, 7)

# Call the add method using the + operator
c = a + b  # This triggers a.__add__(b)

# Access the attributes of the result
print(f"c.a = {c.a}")
print(f"c.b = {c.b}")
print(f"c.c = {c.c}")
```
