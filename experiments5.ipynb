```python
#Write a program to count and display the number of capital letters in a given string.

string = input("Enter a string: ")
capital_count = sum(1 for char in string if char.isupper())
print(f"The number of capital letters in the string is: {capital_count}")


```

```python
#2.	Count total number of vowels in a given string

string = input("Enter a string: ")
num = 0
for word in string:
    if word in 'aeiouAEIOU':
        num += 1
print(f"Total number of vowels: {num}")
```

```python
#3.	Input a sentence and print words in separate lines

sentence = input("Enter a sentence: ")
words = sentence.split()
for word in words:
    print(word)
    
```

```python
"""4.	WAP to enter a string and a substring. You have to print the number of times that the substring occurs in the given string. String traversal will take place from left to right, not from right to left. 
Sample Input 
ABCDCDC 
CDC 
Sample Output 
2 
"""

string = input("Enter a string: ")
print(f"The string is: {string}")
substring = input("Enter a substring: ")
count = string.count(substring)
print(f"The substring '{substring}' occurs {count} times in the string '{string}'.")
```

```python
"""5.	Given a string containing both upper and lower-case alphabets. Write a Python program to count the number of occurrences of each alphabet (case insensitive) and display the same. 
Sample Input 
ABaBCbGc 
Sample Output 
2A 
3B 
2C 
1G 
"""

string = input("Enter a string: ")
string = string.upper()
alphabet_count = {}
for char in string:
    if char.isalpha():
        if char in alphabet_count:
            alphabet_count[char] += 1
        else:
            alphabet_count[char] = 1
for char, count in alphabet_count.items():
    print(f"{count}{char}")

    
```

```python
#6.	Program to count the number of unique words in a given sentence using sets

sentence = input("Enter a sentence: ")
words = sentence.split()
unique_words = set(words)
print(f"The number of unique words in the sentence is: {len(unique_words)}")

```

```python
#7.	Create 2 sets s1 and s2 of n fruits each by taking input from user and find: 

# a.	Union of s1 and s2
n = int(input("Enter the number of fruits for each set: "))
s1 = set()
s2 = set()
print("Enter fruits for set s1:")
for _ in range(n):
    fruit = input("Fruit: ")
    s1.add(fruit)
print("Enter fruits for set s2:")
for _ in range(n):
    fruit = input("Fruit: ")
    s2.add(fruit)
union_set = s1.union(s2)
print(f"Union of s1 and s2: {union_set}")

```

```python
"""8.	Take two sets and apply various set operations on them: 
S1 = {Red, yellow, orange, blue} 
S2 = {violet, blue, purple}
"""

S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}
print(f"Union of S1 and S2: {S1.union(S2)}")
print(f"Intersection of S1 and S2: {S1.intersection(S2)}")
print(f"Difference of S1 and S2 (S1 - S2): {S1.difference(S2)}")
print(f"Difference of S1 and S2 (S2 - S1): {S2.difference(S1)}")
print(f"Symmetric Difference of S1 and S2: {S1.symmetric_difference(S2)}")
```
