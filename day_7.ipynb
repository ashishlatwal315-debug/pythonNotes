```python
#__repr__() is a fallback method that is used when __str__() is not defined,
#  and it returns a string representation of the object.
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student name is {self.name}"

    def __repr__(self):
        return f"Student('{self.name}')"

s = Student("ashish")

print(s)        # uses __str__()
```

```python
class Test:
    def __init__(self, x):
        self.x = x#a.x = 10

    def __eq__(self, other):
        return self.x == other.x

a = Test(10)
b = Test(10)
# a == b call a.__eq__(b)
print(a == b)
```

```python
class Test:
    def __init__(self, x):
        self.x = x

    def __lt__(self, other):
        return self.x < other.x

a = Test(5)
b = Test(10)

print(a < b)
```

```python
# __call__ will give true when the the object can be get called
# false when the object can not be called 
class Add:
    def __init__(self,a, b):
        self.a = a
        self.b = b 
        
    def __call__(self,):
        return self.a + self.b

obj = Add(10,20)
print(obj()) # calling object like function
print(callable(obj))
```
