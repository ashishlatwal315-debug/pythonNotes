```python
"""Q1. Class Variable vs Instance Variable

Create a class Student with:

    one class variable: school_name = "UPES"
    two instance variables: name and age

Create two objects and print:

    school name using both objects
    name and age of both students
"""


class student:
    school_name = "upes"

    def __init__(self,name,age):
        self.name = name 
        self.age = age

s1 = student("asshis",11)
s2 = student("ayush",12)

print("school",s1.school_name)
print("student 1 :",s1.name,s1.age)
print("student 2 :",s2.name,s2.age)
```

```python
"""Q2. Instance Method and Class Method

Create a class Employee with:

    class variable: company = "ABC Ltd"
    instance variables: name, salary
    instance method show_details() to print employee details
    class method change_company(new_name) to change company name for all employees

Create two objects. Change the company name using class method. Print details before and after change."""


class employee:

    company = "ABC Ltd"

    def __init__(self,name,salary):
        self.name = name
        self.salary = salary

    def show_details(self):
        print(f"name:{self.name},salary:{self.salary},company:{employee.company}")
    
    @classmethod
    def change_company(cls,new_name):
        cls.company = new_name

e1= employee("ashish",25000)
e2= employee("ayush",150)

#details before
print("before changing company")

e1.show_details()
e2.show_details()

#changing name

employee.change_company("ashish pvt ltd")


#details after
print("details after changing")

e1.show_details()
e2.show_details()



        

```

```python
"""Q3. Static Method

Create a class MathHelper with a static method:

is_even(num)

It should return True if the number is even, otherwise False.

Test the method with:

    8
    11

Expected Concept

This is based on the syllabus section on static methods as utility methods."""

class mathhelper:

    @staticmethod
    def is_even(num):
        return num%2 == 0
    

print(mathhelper.is_even(8))
print(mathhelper.is_even(11))
```

```python
"""Q4. Public and Private Data Members

Create a class BankAccount with:

    public variable: name
    private variable: __balance

Add methods:

    deposit(amount)
    withdraw(amount)
    show_balance()

Requirements:

    deposit should increase balance
    withdraw should reduce balance only if sufficient balance exists
    if balance is not enough, print "Insufficient balance"

Create one object and test all methods."""


class bankaccount:

    def __init__(self, name, balance):
        self.name = name
        self._balance = balance
    
    def deposit(self, amount):
        self._balance += amount
        print("successfully deposited")
    
    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
            print("successfully withdrawn")
        else:
            print("insufficient balance")

    def show_balance(self):
        print(f"accountholder: {self.name}, balance: {self._balance}")


acc1 = bankaccount("ashish", 5000)

acc1.show_balance()
acc1.deposit(500)
acc1.show_balance()
acc1.withdraw(300)
acc1.show_balance()
```

```python
"""Q6. Magic Methods / Operator Overloading

Create a class Product with:

    attributes: name, price

Implement the following magic methods:

    __init__()
    __str__()
    __add__() → should return sum of prices of two products
    __eq__() → should return True if both products have same price

Create three objects:

p1 = Product("Laptop", 50000)
p2 = Product("Phone", 30000)
p3 = Product("Tablet", 50000)

Then print:

    p1
    p1 + p2
    p1 == p3
    p1 == p2
"""

class product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    
    def __str__(self):
        return f"product:{self.name},price:{self.price}"
    
    def __add__(self, other):
        return self.price+other.price

    def __eq__(self, other):
        return self.price == other.price


p1=product("laptop",50000)
p2=product("speaker",4000)
p3=product("tab",7000)

print(p1)
print(p1+p2)
print(p1==p3)
print(p1==p2)
```
