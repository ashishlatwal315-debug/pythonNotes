```python
class student():
    def __init__(self ,name, age, clas):
# __init__ method - constructor, value initialize 
# self parameter = refrance or connection build btw class and odject
        self.name = name# attributes or properties 
        self.age = age
        self.classs = clas
student1 = student("ashish",18,89)# object
print(student1.name,student1.age,student1.classs)
```

```python
class sport():
    def __init__(self,name,sport,clas):
        self.name = name
        self.sport = sport
        self.clas= clas
    def stu_detail(self):
        print(f"{self.name} of class {self.clas} play {self.sport} ")

data = sport("ashish","cricket","12") 
data2 = sport("amit","football","11")
data.stu_detail()#sport.stu_detail(data)
data2.stu_detail()#sport.stu_detail(data2)
```
