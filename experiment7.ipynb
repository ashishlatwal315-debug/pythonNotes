```python
"""1 1.	Write a Python function to find the maximum and minimum numbers from a sequence of numbers.  (Note: Do not use built-in functions.) """

def find_max_min(numbers):
    if not numbers:
        return None, None  # Return None for both if the list is empty

    max_num = numbers[0]
    min_num = numbers[0]

    for num in numbers:
        if num > max_num:
            max_num = num
        elif num < min_num:
            min_num = num

    return max_num, min_num
# Example usage:
numbers = [3, 1, 4, 1, 5, 9
, 2, 6, 5, 3, 5]
max_number, min_number = find_max_min(numbers)
print(f"Maximum number: {max_number}")
print(f"Minimum number: {min_number}")

```

```python
"""2.	Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number."""

def sum_of_cubes(n):
    if n <= 0:
        return 0  # Return 0 for non-positive integers

    total_sum = 0
    for i in range(1, n):
        total_sum += i ** 3

    return total_sum
# Example usage:
number = 5
result = sum_of_cubes(number)
print(f"The sum of the cubes of all positive integers smaller than {number} is: {result}")
```

```python
"""3.	Write a Python function to print 1 to n using recursion. (Note: Do not use loop) """
def print_numbers(n):
    if n <= 0:
        return  
    print_numbers(n - 1)  
    print(n)  # Print the current number after the recursive call

# Example usage:
n = 5
print(f"Numbers from 1 to {n}:")
print_numbers(n)
```

```python
#4.	Write a recursive function to print Fibonacci series upto n terms

def fibonacci(n, a=0, b=1):
    if n <= 0:
        return
    print(a, end=' ')
    fibonacci(n - 1, b, a + b)
# Example usage:
n_terms = 10
print(f"Fibonacci series up to {n_terms} terms:")
fibonacci(n_terms)
```

```python
#5.	Write a lambda function to find volume of cone. 

volume_of_cone = lambda r, h: (1/3) * 3.14159 * r**2 * h
# Example usage:
radius = float(input("Enter the radius of the cone: "))
height = float(input("enter the height of the cone"))
volume = volume_of_cone(radius, height)
print(f"The volume of the cone with radius {radius} and height {height} is: {volume}")
```

```python
"""6.	Write a lambda function which gives tuple of max and min from a list. 
Sample input: [10, 6, 8, 90, 12, 56] 
Sample output: (90,6) 
"""

max_min_tuple = lambda lst: (max(lst), min(lst)) if lst else (None, None)
# Example usage:
numbers = [10, 6, 8, 90, 12, 56]
result = max_min_tuple(numbers)
print(f"The maximum and minimum numbers from the list {numbers} are: {result}")
```

```python
"""7.	Write functions to explain mentioned concepts: 
a.	Keyword argument 
b.	Default argument 
c.	Variable length argument"""

# a. Keyword argument
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"
# Example usage:
print(greet(name="Alice"))
print(greet(name="Bob", greeting="Hi"))
# b. Default argument
def power(base, exponent=2):
    return base ** exponent
# Example usage:
print(power(3))
print(power(3, 3)) 

def sum_all(*args):
    return sum(args)
# Example usage:
print(sum_all(1, 2, 3))
print(sum_all(4, 5))  
print(sum_all())  

```
