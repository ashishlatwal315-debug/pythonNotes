```python
# # ENCAPSULATION
# Encapsulation is the process of making attributes (data) private OR restrict access to certain atributes  
# and providing access to them through public methods

class student:
    def __init__(self, name,grade,percentage):
        self.name= name
        self.grade= grade
        self.__percentage= percentage#double underscore limit acess / make it private

    def stu(self):
        print(f"{self.name} from grade{self.grade} has {self.percentage}  ")
student1 = student("ashish",12,89)           
student2 = student("amit",11,79) 
student1.stu()
student2.stu()
print(student1.__percentage)# it will show error//'student' object has no attribute 'percentage'
```

```python
# INHERITANCE
# allow one class (child) to reuse property and method of another class (parent)
class student:
    def __init__(self, name,grade,percentage):
        self.name= name
        self.grade= grade
        self.percentage= percentage
    
    def stu(self):
        print(f"{self.name} from grade{self.grade} has{self.percentage+2} ") 
student1 = student("ashish",12,89)           
student2 = student("amit",11,79) 
student1.stu()
student2.stu()


class newstudent(student):
    def __init__(self, name, grade, percentage,strem):
# old parameter for parent class and new parameter for new class
        super().__init__(name, grade, percentage)# we not use strem here because we get it upside 
        self.strem= strem
        

        def stu(self):
            super().stu()# method inheritence from parent class

new = newstudent("dhruv",99,99,"math")
print(new.strem,new.name)
print(student1.grade,new.name)
student1.stu()

new.stu()#  INHERITANCE value of class method but using differnt object to print


```

```python
# # POLYMORPHISM
# Polymorphism is the ability of a function or me thod to perform different tasks
#  depending on the object or data it is used with.

class student:
    def __init__(self, name,grade,percentage):
        self.name= name
        self.grade= grade
        self.percentage= percentage
    
    def stu(self):
        print(f"{self.name} from grade{self.grade} has{self.percentage+2} ") 
student1 = student("ashish",12,89)           

student1.stu()



class newstudent(student):
    def __init__(self, name, grade, percentage,strem):

        super().__init__(name, grade, percentage)
        self.strem= strem
        

    def stu(self):
        super().stu()

new = newstudent("dhruv",99,99,"math")

student1.stu()

new.stu()#   POLYMORPHISM value of class method but using differnt object to print
# we can also use student1,student2 in place of new



```
